```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

metrics = [
    "Precision (not liked)",
    "Precision (liked)",
    "Recall (not liked)",
    "Recall (liked)",
]

cnn_values_pr = [
    0.5974025974025974,
    0.6341463414634146,
    0.5054945054945055,
    0.7155963302752294,
]

logreg_values_pr = [
    0.5816326530612245,
    0.6666666666666666,
    0.6263736263736264,
    0.6238532110091743,
]

df = pd.DataFrame({
    "Metric": metrics,
    "CNN (scalogram)": cnn_values_pr,
    "Logistic Regression (bandpower)": logreg_values_pr
})
print(df)

accuracy_metrics = ["Accuracy"]
cnn_acc = [0.62]
logreg_acc = [0.625]
rand_acc = [.465]

x = np.arange(len(accuracy_metrics))
width = 0.25

plt.figure(figsize=(6, 5))
bars_cnn = plt.bar(x - width, cnn_acc, width, label="CNN (scalogram)")
bars_log = plt.bar(x, logreg_acc, width, label="Logistic Regression (bandpower)")
bars_rand = plt.bar(x + width, rand_acc, width, label="Random baseline")

plt.xticks(x, accuracy_metrics)
plt.ylabel("Score")
plt.ylim(0, 1.05)
plt.title("Accuracy Comparison\nCNN (scalogram) vs Logistic Regression (bandpower) vs Random")
plt.legend()
plt.tight_layout()
```
